```python
file_path = 'input'
with open(file_path, 'r') as file:
    lines = file.readlines()
    ranges_temp = lines[0].strip().split(',')

parts = []
for part in ranges_temp:
    start = int(part.split('-')[0])
    end = int(part.split('-')[1])
    parts.append([start, end])

total = 0

for part in parts:
    begin = part[0]
    end = part[1]
    for i in range(begin, end+1):
        i_str = str(i)
        if len(i_str)%2 != 0:
            continue
        half_len = len(i_str)//2
        wrong = False
        for j in range(half_len):
            if i_str[j] != i_str[half_len+j]:
                wrong = True
                break
        if wrong is False:
            total += i

print("Total", total)
```
