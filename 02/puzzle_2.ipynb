```python
from textwrap import wrap
from itertools import groupby


file_path = 'input'
with open(file_path, 'r') as file:
    lines = file.readlines()
    ranges_temp = lines[0].strip().split(',')

parts = []
for part in ranges_temp:
    start = int(part.split('-')[0])
    end = int(part.split('-')[1])
    parts.append([start, end])

total = 0

def all_equal(iterable):
    g = groupby(iterable)
    return next(g, True) and not next(g, False)

for part in parts:
    begin = part[0]
    end = part[1]
    for i in range(begin, end+1):
        i_str = str(i)
        for chunk_len in range(1, len(i_str)//2+1):
            chunks = wrap(i_str, chunk_len)
            if all_equal(chunks):
                total += i
                break

print("Total", total)
```
