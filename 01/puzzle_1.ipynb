```python
file_path = 'input'
with open(file_path, 'r') as file:
    lines = file.readlines()

pos = 50
count = 0

for line in lines:
    dir = line[0]
    val = int(line[1:])
    if dir == 'L':
        pos = (pos-val)%100
    if dir == 'R':
        pos = (pos+val)%100
    if pos == 0:
        count += 1

print(count)
```
