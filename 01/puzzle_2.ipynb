```python
file_path = 'input'
with open(file_path, 'r') as file:
    lines = file.readlines()

pos = 50
count = 0

for line in lines:
    dir = line[0]
    val = int(line[1:])

    if dir == 'L':
        while val > 0:
            pos -= 1
            val -= 1
            if pos == 0:
                count += 1
            if pos == -1:
                pos = 99
            
    if dir == 'R':
        while val > 0:
            pos += 1
            val -= 1
            if pos == 100:
                pos = 0
            if pos == 0:
                count += 1


print(count)
```
