```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

rolls = set()

max_x = 0
max_y = 0

for y, line in enumerate(lines):
    line_temp = line.strip().split()
    max_x = len(line_temp[0])-1
    for x, field in enumerate(line_temp[0]):
        if field == '@':
            rolls.add((x, y))
    max_y = y

accessible_rolls = 0

visited = set()
border_rolls = set()

start_x = 0
start_y = 0

for roll in rolls:
    neighbours = 0
    for x_delta in [-1, 0, 1]:
        for y_delta in [-1, 0, 1]:
            if x_delta == 0 and y_delta == 0:
                continue
            roll_x = roll[0] + x_delta
            roll_y = roll[1] + y_delta
            if (roll_x, roll_y) in rolls:
                neighbours += 1
    if neighbours < 4:
        accessible_rolls += 1

print(accessible_rolls)
```
