```python
from itertools import combinations
s
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

points = []
segments = []

for y, line in enumerate(lines):
    line_temp = line.strip().split(",")
    points.append((int(line_temp[0]), int(line_temp[1])))

for i in range(len(points)-1):
    segments.append([points[i], points[i+1]])
segments.append([points[-1], points[0]])

rectangles = []
for pair in combinations(points, 2):
    rectangles.append((pair[0], pair[1]))

rectangles = sorted(rectangles, key=lambda pair: (abs(pair[1][0] - pair[0][0])+1) * (abs(pair[1][1] - pair[0][1])+1), reverse=True)

for rectangle in rectangles:
    for segment in segments:
        crossed = True
        if segment[0][0] <= min(rectangle[0][0], rectangle[1][0]) and segment[1][0] <= min(rectangle[0][0], rectangle[1][0]):
            crossed = False
        if segment[0][0] >= max(rectangle[0][0], rectangle[1][0]) and segment[1][0] >= max(rectangle[0][0], rectangle[1][0]):
            crossed = False
        if segment[0][1] <= min(rectangle[0][1], rectangle[1][1]) and segment[1][1] <= min(rectangle[0][1], rectangle[1][1]):
            crossed = False
        if segment[0][1] >= max(rectangle[0][1], rectangle[1][1]) and segment[1][1] >= max(rectangle[0][1], rectangle[1][1]):
            crossed = False
        
        if crossed:
            break

    if crossed is True:
        continue

    if crossed is False:
        print((abs(rectangle[1][0] - rectangle[0][0])+1) * (abs(rectangle[1][1] - rectangle[0][1])+1))
        break
```
