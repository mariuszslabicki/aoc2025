```python
from itertools import combinations

file_path = 'example_input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

points = []

for y, line in enumerate(lines):
    line_temp = line.strip().split(",")
    points.append((int(line_temp[0]), int(line_temp[1])))

max_area = -1

for pair in combinations(points, 2):
    exclude = False
    for corner in points:
        if corner in pair:
            continue
        if min(pair[0][0], pair[1][0]) <= corner[0] and corner[0] < max(pair[0][0], pair[1][0]) and \
              min(pair[0][1], pair[1][1]) <= corner[1] and corner[1] < max(pair[0][1], pair[1][1]):
            exclude = True
            break
    if exclude:
        continue
    temp_area = (abs(pair[1][0] - pair[0][0])+1) * (abs(pair[1][1] - pair[0][1])+1)
    if temp_area > max_area:
        max_area = temp_area

print(max_area)

# 4566128621 is too high
```
