```python
from itertools import combinations
import math

file_path = 'example_input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

points = []
segments = []

for y, line in enumerate(lines):
    line_temp = line.strip().split(",")
    points.append((int(line_temp[0]), int(line_temp[1])))

for i in range(len(points)-1):
    segments.append([points[i], points[i+1]])
segments.append([points[-1], points[0]])

rectangles = []
for pair in combinations(points, 2):
    rectangles.append((pair[0], pair[1]))

rectangles = sorted(rectangles, key=lambda pair: (abs(pair[1][0] - pair[0][0])+1) * (abs(pair[1][1] - pair[0][1])+1), reverse=True)

for rectangle in rectangles:
    to_the_right = False
    to_the_left = False
    under = False
    over = False
    drop_rectangle = False
    for segment in segments:
        if segment[0][0] == segment[1][0]:
            seg_x = segment[0][0]
            if seg_x < min(rectangle[0][0], rectangle[1][0]):
                to_the_left = True
            if seg_x > max(rectangle[0][0], rectangle[1][0]):
                to_the_right = True

        if segment[0][1] == segment[1][1]:
            seg_y = segment[1][0]
            if seg_y < min(rectangle[0][1], rectangle[1][1]):
                under = True
            if seg_y > max(rectangle[0][1], rectangle[1][1]):
                over = True
        if to_the_left is True or to_the_right is True and \
            under is True or over is True:
            drop_rectangle = True
            continue
        else:
            
            break

    if drop_rectangle is False:
        print("Znalazlem")
        continue
    else:
        print(rectangle, segment)
```
