```python
import numpy as np

file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

gifts = []
gifts_areas = []
floorplans = []
requirements_set = []


gift = []
area = 0
for y, line in enumerate(lines):
    line_temp = line.strip().split(" ")
    if line_temp == ['']:
        gift = np.array(gift)
        gifts.append(gift)
        gifts_areas.append(area)
        gift = []
        area = 0
        continue
    if len(line_temp) == 1 and line_temp[0][-1] != ':':
        row = []
        
        for element in line_temp[0]:
            if element == '.':
                row.append(0)
            if element == '#':
                row.append(1)
                area += 1
        gift.append(row)
    if len(line_temp) > 1:
        size = line_temp[0][0:-1].split("x")
        floorplan = (int(size[0]), int(size[1]))
        floorplans.append(floorplan)
        requirements = [int(i) for i in line_temp[1:]]
        requirements_set.append(requirements)


fits_counter = 0

for i, requirements in enumerate(requirements_set):
    total_area = 0
    for j in range(len(requirements)):
        total_area += requirements[j] * gifts_areas[j]
    if total_area > floorplans[i][0]*floorplans[i][1]:
        pass
    else:
        fits_counter += 1

print(fits_counter)
```
