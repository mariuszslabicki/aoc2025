```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

rolls = set()

max_x = 0
max_y = 0

start = None
splitters = set()

for y, line in enumerate(lines):
    line_temp = line.strip().split()
    max_x = len(line_temp[0])-1
    for x, field in enumerate(line_temp[0]):
        if field == 'S':
            start = (x, y)
        if field == '^':
            splitters.add((x, y))
    max_y = y

beams = dict()
beams[start] = 1

while len(beams) > 0:
    new_beams = dict()
    for beam in beams:
        prev_count = beams[beam[0], beam[1]]
        new_y = beam[1] + 1
        if new_y > max_y:
            break

        new_x = beam[0]
        if (new_x, new_y) not in splitters:
            if new_x < 0 or new_x > max_x:
                continue

            if (new_x, new_y) not in new_beams:
                new_beams[(new_x, new_y)] = prev_count
            else:
                new_beams[(new_x, new_y)] = prev_count + new_beams[(new_x, new_y)]

        else:
            for new_x in [beam[0]-1, beam[0]+1]:
                if new_x < 0 or new_x > max_x:
                    continue
                if (new_x, new_y) not in new_beams:
                    new_beams[(new_x, new_y)] = prev_count
                else:
                    new_beams[(new_x, new_y)] = prev_count + new_beams[(new_x, new_y)]

    if len(new_beams) == 0:
        break
    beams = new_beams

total = 0
for beam in beams:
    total += beams[beam]

print(total)
```
