```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

rolls = set()

max_x = 0
max_y = 0

start = None
splitters = set()

for y, line in enumerate(lines):
    line_temp = line.strip().split()
    max_x = len(line_temp[0])-1
    for x, field in enumerate(line_temp[0]):
        if field == 'S':
            start = (x, y)
        if field == '^':
            splitters.add((x, y))
    max_y = y

beams = set()
beams.add(start)

splits = 0

while len(beams) > 0:
    new_beams = set()
    for beam in beams:
        next_steps = []
        if (beam[0], beam[1]+1) in splitters:
            splits += 1
            next_steps.append((beam[0]-1, beam[1]+1))
            next_steps.append((beam[0]+1, beam[1]+1))
        else:
            next_steps.append((beam[0], beam[1]+1))
        for next_step in next_steps:
            if next_step[0] < 0 or next_step[0] > max_x:
                continue
            if next_step[1] < 0 or next_step[1] > max_y:
                continue
            new_beams.add(next_step)
    beams = new_beams

print(splits)
```
