```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

values = []
all_shifts = []

for y, line in enumerate(lines):
    line_temp = line.strip().split(" ")

    val_raw = list((line_temp[0]))[1:-1]
    val = 0
    for i, c in enumerate(val_raw):
        if c == '#':
            val += pow(2, i)

    shifts_raw = line_temp[1:-1]

    shifts = []
    for one_shift in shifts_raw:
        one_shift_int = 0
        for i in one_shift[1:-1].split(','):
            one_shift_int += pow(2, int(i))
        shifts.append(one_shift_int)
    values.append(val)
    all_shifts.append(shifts)

total = 0

for i in range(len(values)):
    current_values = set()
    current_values.add(values[i])
    steps_cnt = 0
    while True:
        steps_cnt += 1
        temp_values = set()
        for value in current_values:
            for shift in all_shifts[i]:
                temp_values.add(value ^ shift)
        if 0 in temp_values:
            total += steps_cnt
            break
        current_values = temp_values

print(total)
```
