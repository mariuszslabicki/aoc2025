```python
from itertools import combinations
import numpy as np
import scipy.optimize

file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

values = []
all_shifts = []
buttons = []
expected_counters = []

for y, line in enumerate(lines):
    line_temp = line.strip().split(" ")

    counters_raw = line_temp[-1]
    counters = []
    for one_counter in counters_raw[1:-1].split(','):
        counters.append(int(one_counter))

    buts_raw = line_temp[1:-1]
    buts = []
    for one_but_raw in buts_raw:
        one_but = []
        for i in one_but_raw[1:-1].split(','):
            one_but.append(int(i))
        buts.append(one_but)
    buttons.append(buts)

    expected_counters.append(counters)

total = 0

for i, jolts in enumerate(expected_counters):
    buts = buttons[i]
    A = [[0 for i_ in range(len(buts))] for j in range(len(jolts))]
    for j, but in enumerate(buts):
        for light in but:
            A[light][j] = 1

    c = [1 for i_ in range(len(buts))]
    res = scipy.optimize.linprog(c, A_eq=A, b_eq=jolts, integrality=1)

    total += sum(res.x)

print(int(total))

```
