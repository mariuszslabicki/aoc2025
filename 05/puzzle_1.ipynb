```python

from itertools import combinations

file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

ranges = []
values = []

for y, line in enumerate(lines):
    line_divided = line.strip().split("-")
    if len(line_divided) == 2:
        ranges.append([int(line_divided[0]), int(line_divided[1])])
    if len(line_divided) == 1 and line_divided[0] != '':
        values.append(int(line_divided[0]))

while True:
    updated = False
    for pair in combinations(ranges, 2):
        if pair[1][0] <= pair[0][0] and pair[0][0] <= pair[1][1] or pair[1][0] <= pair[0][1] and pair[0][1] <= pair[1][1]:
            min_val = min(pair[0][0], pair[0][1], pair[1][0], pair[1][1])
            max_val = max(pair[0][0], pair[0][1], pair[1][0], pair[1][1])
            ranges.remove(pair[0])
            ranges.remove(pair[1])
            ranges.append([min_val, max_val])
            updated = True
            break
    if updated is False:
        break

found = 0

for value in values:
    for range in ranges:
        if range[0] <= value and value <= range[1]:
            found += 1
            break
print(found)
```
