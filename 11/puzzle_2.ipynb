```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

nodes = dict()

for y, line in enumerate(lines):
    node = line.strip().split(":")[0].strip()
    outs = line.strip().split(":")[1].strip().split(" ")
    outs_list = [out for out in outs]
    nodes[node] = outs

def count_paths(start, stop, exclusion_ranges):
    no_of_paths = 0

    to_exclude = set()
    for exclusion_range in exclusion_ranges:
        reduce_wave = [exclusion_range[0]]
        while len(reduce_wave) > 0:
            new_wave = []
            for end in reduce_wave:
                for exit in nodes[end]:
                    if exit == exclusion_range[1] or exit in to_exclude:
                        continue
                    to_exclude.add(exit)
                    new_wave.append(exit)
            reduce_wave = new_wave

    wave = [start]
    while len(wave) > 0:
        new_wave = []
        for end in wave:
            for exit in nodes[end]:
                if exit in to_exclude:
                    continue
                if exit == stop:
                    no_of_paths += 1
                    continue
                new_wave.append(exit)
        wave = new_wave
    return no_of_paths

no_of_paths_1a = count_paths('svr', 'fft', [['fft', 'out']])
# print(no_of_paths_1a)
no_of_paths_1b = count_paths('fft', 'dac', [['dac', 'out']])
# print(no_of_paths_1b)
no_of_paths_1c = count_paths('dac', 'out', [])
# print(no_of_paths_1c)

print(no_of_paths_1a*no_of_paths_1b*no_of_paths_1c)
```
