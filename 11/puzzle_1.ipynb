```python
# %%timeit
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

nodes = dict()

for y, line in enumerate(lines):
    node = line.strip().split(":")[0].strip()
    outs = line.strip().split(":")[1].strip().split(" ")
    outs_list = [out for out in outs]
    nodes[node] = outs

# print(nodes)

wave = ['you']

no_of_paths = 0

while len(wave) > 0:
    new_wave = []
    for end in wave:
        for exit in nodes[end]:
            if exit == 'out':
                no_of_paths += 1
                continue
            new_wave.append(exit)
    wave = new_wave

print(no_of_paths)
```
