```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

for line in lines:
    line_temp = line.strip().split()
    line_divided = []
    for element in line_temp[0]:
        line_divided.append(int(element))
    sequences.append(line_divided)

total = 0

for sequence in sequences:
    for i in range(99, -1, -1):
        part_a = i//10
        part_b = i%10

        try:
            pos_a = sequence.index(part_a)
        except ValueError:
            continue
        try:
            pos_b = sequence[pos_a+1:].index(part_b)
        except ValueError:
            continue

        total += 10*sequence[pos_a] + sequence[pos_a+1+pos_b]
        break

print(total)
```
