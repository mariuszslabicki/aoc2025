```python
file_path = 'input'

sequences = []

with open(file_path, 'r') as file:
    lines = file.readlines()

values = []
operations = []

for y, line in enumerate(lines):
    line_divided = line.strip().split()
    if line_divided[0] not in ['+', '*']:
        row = []
        for element in line_divided:
            row.append(int(element))
        values.append(row)
    if line_divided[0] in ['+', '*']:
        operations = line_divided

total = 0

for i, op in enumerate(operations):
    if operations[i] == '+':
        agg_val = 0
    if operations[i] == '*':
        agg_val = 1
    for line in values:
        if operations[i] == '+':
            agg_val += line[i]
        if operations[i] == '*':
            agg_val *= line[i]
    total += agg_val

print(total)
```
